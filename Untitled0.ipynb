```python
import pandas as pd

df = pd.read_csv("Student Mental health.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

df = df.drop_duplicates()

print("Data cleaned successfully")
```

```python
df.to_csv("cleaned_student_mental_health.csv", index=False)
```

```python
import matplotlib.pyplot as plt

df["Choose your gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
import matplotlib.pyplot as plt

df["Do you have Depression?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Depression Analysis")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt

df["Do you have Anxiety?"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Anxiety Analysis")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt

df["Age"].hist(bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")

plt.show()
```

```python
df["What is your CGPA?"].value_counts().plot(
    kind="bar"
)

plt.title("CGPA Distribution")
plt.xlabel("CGPA Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
print("Conclusion")
print("1. Dataset was cleaned by handling missing values and duplicates.")
print("2. Gender distribution was analyzed.")
print("3. Depression and anxiety levels were visualized.")
print("4. Age distribution of students was studied.")
print("5. CGPA distribution was analyzed.")
```

```python
df.to_csv("cleaned_student_mental_health.csv", index=False)

print("Cleaned dataset saved")
```
